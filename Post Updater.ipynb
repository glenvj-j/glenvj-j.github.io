```python
def generate_html(projects, custom_order=None):
    html_content = ""
    row_content = ""

    # Start with custom ordered projects
    ordered_keys = custom_order or []
    ordered_projects = []

    seen_keys = set()

    # Add custom-ordered projects first
    for key in ordered_keys:
        if key in projects:
            ordered_projects.append((key, projects[key]))
            seen_keys.add(key)

    # Add the rest of the projects that are not in the custom order
    for key, project in projects.items():
        if key not in seen_keys:
            ordered_projects.append((key, project))

    total_projects = len(ordered_projects)

    for idx, (key, project) in enumerate(ordered_projects, start=1):
        project_html = f"""
        <div class="col-sm-4">
            <a href="{project['highlight']}" title="" class="black-image-project-hover" target="_blank" rel="noopener noreferrer">
                <img src="./assets/images/{project['image']}" alt="" class="img-responsive">
            </a>
            <div class="card-container card-container-lg">
                <h4>{str(idx).zfill(3)}/{str(total_projects).zfill(3)}</h4>
                <h3>{project['title']}</h3>
                <p>{project['description']}</p>
        """
        for link_text, link_url in project['links']:
            project_html += f'<a href="{link_url}" title="" class="btn btn-default" target="_blank" rel="noopener noreferrer">{link_text}</a> '
        
        project_html += """
            </div>
        </div>
        """

        row_content += project_html

        # Group every 6 projects into a row
        if idx % 3 == 0 or idx == total_projects:
            html_content += f"<div class='row'>{row_content}</div>"
            row_content = ""

    return html_content
```

```python
custom_order = ['CivDexAI','Olist', 'Passport','Employee','Taxi', 'Dukcapil']
html_output = generate_html(projects, custom_order)
print(html_output)
```
